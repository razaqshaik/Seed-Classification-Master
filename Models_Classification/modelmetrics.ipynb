```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Step 1: Define the metrics for each model
data = {
    'Model': [
        'CNN', 'DenseNet', 'EfficientNet', 'InceptionV3', 
        'MobileNetV2', 'NASNet', 'ResNet', 'VGG19'
    ],
    'Accuracy': [81, 88, 86, 92, 89, 89, 88, 89],
    'Macro Avg Precision': [80, 89, 87, 91, 89, 89, 89, 89],
    'Macro Avg Recall': [80, 88, 86, 90, 88, 88, 88, 89],
    'Macro Avg F1-Score': [80, 88, 89, 90, 88, 88, 88, 89]
}

# Step 2: Convert to DataFrame
df = pd.DataFrame(data)

# Step 3: Plot the metrics
metrics = ['Accuracy', 'Macro Avg Precision', 'Macro Avg Recall', 'Macro Avg F1-Score']
x = np.arange(len(df['Model']))  # Model positions

# Plot
fig, ax = plt.subplots(figsize=(16, 10))  # Increase figure size
bar_width = 0.15  # Reduce bar width for better spacing

# Create bars for each metric
for i, metric in enumerate(metrics):
    ax.bar(x + i * bar_width, df[metric], width=bar_width, label=metric)

# Customizations
ax.set_xlabel('Models', fontsize=14)
ax.set_ylabel('Metrics (%)', fontsize=14)
ax.set_title('Model Metrics Comparison', fontsize=18)
ax.set_xticks(x + bar_width * (len(metrics) - 1) / 2)
ax.set_xticklabels(df['Model'], fontsize=12, rotation=45)
ax.legend(fontsize=12, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)  # Move legend outside
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()

```
